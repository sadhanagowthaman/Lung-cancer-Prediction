```python
# Import the neccessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Dataset.
data = pd.read_csv('Dataset.csv')
```

```python
data.info()
```

```python
data.columns
```

```python
df = data.rename({'ALCOHOL CONSUMING':'Alcohol_Consuming','CHRONIC DISEASE':'Chronic_disease','SHORTNESS OF BREATH':'Shortness_of_breath',
          'SWALLOWING DIFFICULTY':'Swallowing_difficulty','CHEST PAIN':'chest_pain','GENDER':'Gender','AGE':'Age','SMOKING':'Smoking',
          'YELLOW_FINGERS':'Yellow_fingers','ANXIETY':'Anxiety','PEER_PRESSURE':'Peer_pressure','FATIGUE ':'Fatigue','ALLERGY ':'Allergy',
          'WHEEZING':'Wheezing','COUGHING':'Coughing','LUNG_CANCER':'Lung_cancer'},axis=1)
```

```python
df.head()
```

```python
#Histogram
df['Smoking'].hist(figsize=(10,5), color='y', alpha=1)
plt.xlabel('Smoking rate')
plt.ylabel('count')
plt.title('Total smoke rate')
```

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
plt.hist(df['Smoking'])
plt.title("Average smoke rate")
plt.xlabel("For average smoking")
plt.ylabel("Total Users in our dataset")
plt.subplot(1,2,2)
plt.hist(df['Alcohol_Consuming'])
plt.xlabel('Alcohol')
plt.title('Alcohol consuming')
plt.ylabel("Number of Users")
plt.show()
```

```python
# Histogram of the project.
df.hist(figsize=(15,55), color='g',layout=(15,4))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.stripplot(y = df['Alcohol_Consuming'], x = df['Age'], ax=ax)
plt.title("Age wise alcohol consuming")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25,12))
sns.heatmap(df.corr(),annot = True, fmt='0.2%',cmap = 'flare',ax=ax)
#YlGnBu, Reds, tab10, rocket, mako, flare, crest, magma, viridis, cubehelix, YlOrBr, vlag, icefire, Spectral, coolwarm, Blues
```
